```python
df = df.rename(columns={"Qchat-10-Score": "total_score"})
df = df.rename(str.lower, axis='columns')
df.head()
```

```python
df['sex'] =pd.get_dummies(df['sex'],drop_first=True)
df['jaundice'] =pd.get_dummies(df['jaundice'],drop_first=True)
df['family_mem_with_asd'] =pd.get_dummies(df['family_mem_with_asd'],drop_first=True)
df['within24_36']= pd.get_dummies(df['age_mons']>24,drop_first=True)
df['within0_12'] = pd.get_dummies(df['age_mons']<13,drop_first=True)
df['class'] =pd.get_dummies(df['class '],drop_first=True)
```

```python
df['assigner'] = df["assigner"].str.lower()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['ethnicity'] = le.fit_transform(df.ethnicity)
df['assigner'] = le.fit_transform(df.assigner)
```

```python
df = df.drop(['total_score', 'class ', 'case_no','age_mons'],axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X= df.iloc[:,:-1]
y= df.iloc[:,-1]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier(n_estimators=500)
rfc.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
pred_rfc= rfc.predict(X_test)
print(confusion_matrix(y_test,pred_rfc))
print(classification_report(y_test,pred_rfc))
```

```python
import pickle
pkl_model = 'pkl_model.pkl'
pickle.dump(rfc, open(pkl_model, 'wb'))
```
